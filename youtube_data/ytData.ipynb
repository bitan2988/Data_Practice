```python
categorySet = set()
trendingCategoryCount = {}
categoryMetadata = {} #trendingCategoryViewsLikesDislikesComments

for i in range(len(data)):
    categoryId = data.iloc[i]['category_id']
    categorySet.add(categoryId)
    if categoryId in trendingCategoryCount:
        trendingCategoryCount[categoryId] += 1
        views = data.iloc[i]['views'] + categoryMetadata[categoryId][0]
        likes = data.iloc[i]['likes'] + categoryMetadata[categoryId][1]
        dislikes = data.iloc[i]['dislikes'] + categoryMetadata[categoryId][2]
        comments = data.iloc[i]['comment_count'] + categoryMetadata[categoryId][3]
        categoryMetadata[categoryId] = [views, likes, dislikes, comments]
        
    else:
        trendingCategoryCount[categoryId] = 1
        likes = data.iloc[i]['likes']
        views = data.iloc[i]['views']
        dislikes = data.iloc[i]['dislikes']
        comments = data.iloc[i]['comment_count']
        categoryMetadata[categoryId] = [views, likes, dislikes, comments]

```

```python
print(categorySet)
print(trendingCategoryCount)
print(sorted(trendingCategoryCount.values(), reverse=True))

print('\nCorrect manner of sorting in accordance to value pair')
sortedData = sorted(trendingCategoryCount.items(), reverse=True, key=lambda keyVal: (keyVal[1], keyVal[0]))
print(sortedData)
```

```python
print(f'topCategoryId = {sortedData[0][0]}, totalTimesTrending = {sortedData[0][1]}')
print('Metdata')
print(f'    Total Views : {categoryMetadata[sortedData[0][0]][0]}')
print(f'    Total Likes : {categoryMetadata[sortedData[0][0]][1]}')
print(f'    Total Dislikes : {categoryMetadata[sortedData[0][0]][2]}')
print(f'    Total Comments : {categoryMetadata[sortedData[0][0]][3]}')
```
